## Recursive discretization

This function takes a computation point, a tesseroid, and a distance/size ratio and recursively breaks the tesseroid into smaller ones until `distance > size*ratio` for every smaller tesseroid. This is what happens in the tesseroid computations automatically.

```python
def discretize(tesseroid, point, ratio):
    d2r = np.pi/180
    lon, lat, h = point
    lon *= d2r
    lat *= d2r
    r = h + MEAN_EARTH_RADIUS
    tr = tesseroid.top + MEAN_EARTH_RADIUS
    tlat = d2r*0.5*(tesseroid.s + tesseroid.n)
    tlon = d2r*0.5*(tesseroid.w + tesseroid.e)
    cospsi = np.sin(lat)*np.sin(tlat) + np.cos(lat)*np.cos(tlat)*np.cos(lon - tlon)
    distance = np.sqrt(r**2 + tr**2 - 2*r*tr*cospsi)
    size = max([MEAN_EARTH_RADIUS*d2r*(tesseroid.e - tesseroid.w),
                MEAN_EARTH_RADIUS*d2r*(tesseroid.n - tesseroid.s),
                tesseroid.top - tesseroid.bottom])
    if distance > size*ratio:
        return [tesseroid]
    else:
        return [t for tess in tesseroid.split(2, 2, 2) for t in discretize(tess, point, ratio)]
```

I'll make the figures with the tesseroid and computation point defined below.

```python
tesseroid = Tesseroid(-55, -25, -80, -70, 500000, 0)
point = [-40, -70, 800000]
```

This function sets the camera position for the 3D plot. This way, all plots will be from the same angle.

```python
def setview(scene):
    scene.scene.camera.position = [16416679.707178073, -37375463.056315556, -8217239.6229258822]
    scene.scene.camera.focal_point = [17524.775500096133, 2455801.3439532616, -6133960.4172555953]
    scene.scene.camera.view_angle = 2.0615843020800004
    scene.scene.camera.view_up = [-0.010129246170078175, 0.048069096295061252, -0.99879265133129702]
    scene.scene.camera.clipping_range = [23552679.71500038, 63396872.424586065]
    scene.scene.camera.compute_view_plane_normal()
    scene.scene.render()
```

Plot the original tesseroid.

```python
scene = myv.figure(zdown=False, size=(800, 600))
myv.earth(opacity=1, color=(0.7, 0.7, 0.7))
myv.continents(linewidth=2)
# Split the tesseroid so that it doesn't look so square in the plot
myv.tesseroids(tesseroid.split(35, 35, 20), edges=False, color=(0.4, 0.4, 0.4))
myv.points([point], size=30000, spherical=True)
myv.meridians(range(0, 360, 30))
myv.parallels(range(-90, 0, 15))
setview(scene)
myv.savefig('../figs/tesseroid-split-whole.png')
myv.show()
```

Now make plots of the discretization for various distance/size ratios.

```python
for split in [0.5, 1, 2]:
    scene = myv.figure(zdown=False, size=(800, 600))
    myv.earth(opacity=1, color=(0.7, 0.7, 0.7))
    myv.continents(linewidth=2)
    model = discretize(tesseroid, point, split)
    print "Number of tesseroids:", len(model)
    myv.tesseroids(model, linewidth=2, edges=True, color=(0.4, 0.4, 0.4))
    myv.points([point], size=30000, spherical=True)
    myv.meridians(range(0, 360, 30))
    myv.parallels(range(-90, 0, 15))
    setview(scene)
    myv.savefig('../figs/tesseroid-split-ratio%g.png' % (split))
myv.show()
```
