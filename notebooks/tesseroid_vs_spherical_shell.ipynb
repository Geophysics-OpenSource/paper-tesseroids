# Tesseroid versus spherical shell

We want to analyse the error involed in calculating the gravitational potential and its derivatives using tesseroids.
The fields are calculated using a Gauss-Legendre Quadrature (GLQ) integration.
Ku (1977) showed that the error of the GLQ for rectangular prisms depends on the ratio of the distance between GLQ discretization points (nodes) and the distance between the prism and the observation point.

## Tesseroid models

We will need to make tesseroid models of half a spherical shell for varying tesseroid sizes.
The shell thickness is set to 50 km to represent a crustal model.


```python
def make_model(size, thickness=50000, density=2700):
    # Stream the model to avoid overloading the memory
    for w in np.arange(0, 180, size):
        for s in np.arange(-90, 90, size):
            yield Tesseroid(w, w+size, s, s+size, 0, -thickness, {'density':density})

def dump_model(model, fname):
    with open(fname, 'w') as f:
        for t in model:
            f.write('%.10f %.10f %.10f %.10f %.10f %.10f %.10f\n'
                    % (t.w, t.e, t.s, t.n, t.top, t.bottom, t.props['density']))
```

## Spherical shell

First, we need to calculate the effect of a spherical shell.

We'll use only half of a full spherical shell to speed up the computations.

The analytical solution exists for an observation point located on the pole.
Since we're using half a shell, we slice it at the 0 longitude meridian (going through the poles) and pretend that the point is actually on the equator.
This way, we avoid the numerical singularities of tesseroids in the pole.

```python
scene = myv.figure(size=(800, 800), zdown=False)
# The thickness is exagerated for better viewing
surf = myv.tesseroids(make_model(1, thickness=1000000), edges=False)
surf.actor.mapper.scalar_visibility = False
surf.actor.property.color = (0.7, 0.7, 0.7)
myv.earth(opacity=0.2, color=(0.4, 0.4, 0.4))
myv.continents(linewidth=2)
myv.meridians(range(0, 360, 45))
myv.parallels(range(-90, 90, 45))
# Set the view (used the record function in Mayavi to generate the code)
scene.scene.camera.position = [26426614.216552604, -11860292.356845967, 3398148.0172464354]
scene.scene.camera.focal_point = [-105672.8039009934, -268668.85375889373, -437.36630568103283]
scene.scene.camera.view_angle = 30.0
scene.scene.camera.view_up = [-0.10402892806057, 0.053073166228378141, 0.99315719861110807]
scene.scene.camera.clipping_range = [10794051.921816513, 52357048.183602169]
scene.scene.camera.compute_view_plane_normal()
scene.scene.render()
myv.savefig('spherical-shell.png')
myv.show()
```

The potential of such a shell is:

$V(r) = 2\pi G \rho \left[ \dfrac{l^3 + {r'}^3}{3r} - 0.5 {r'}^2 \right] \Biggr\rvert_{r'=r_1}^{r'=r_2} $

where $\rho$ is the dentity, $l = \sqrt{r^2 + {r'}^2}$, $r$ is the radius coordinate of the observation point, $r_1$ and $r_2$ are the bottom and top of the spherical shell, respectively.

The $g_z$ component of the gravitational attraction for this shell is the radial derivative of the potential.

The sign should be inverted because our convention is that z points down (toward the center of the Earth)

$g_z(r) = -\dfrac{\partial V}{\partial r} = 2\pi G \rho \left[ \dfrac{l^3 + {r'}^3}{3r^2} - l \right] \Biggr\rvert_{r'=r_1}^{r'=r_2}$

The gravity gradient tensor are the second derivatives of $V$:

$g_{xx}(r) = -\dfrac{1}{2} g_{zz}$

$g_{xy}(r) = 0$

$g_{xz}(r) = 0$

$g_{yy}(r) = -\dfrac{1}{2} g_{zz}$

$g_{yz}(r) = 0$

$g_{zz}(r) = 2\pi G \rho \left[ 2\dfrac{l^3 + {r'}^3}{3r^3} - \dfrac{l}{r} + \dfrac{r}{l} \right] \Biggr\rvert_{r'=r_1}^{r'=r_2}$

Create the tesseroid models and compare the results obtained using Tesseroids for various distance/size ratios.

The results will be saved to a JSON file.

```python
sizes = [0.1, 1, 4, 10, 20] # in degrees
ratios = range(9)
heights = np.array([1, 100, 1000, 10000, 100000, 250000, 500000]) # in meters
lons = 90*np.ones_like(heights)
lats = np.zeros_like(heights)
thickness = 50000
density = 2700
fields = ['pot', 'gz', 'gxx', 'gxy', 'gxz', 'gyy', 'gyz', 'gzz']
#fields = ['gz', 'gzz']
#sizes = [10, 20]
#heights = np.array([250000, 500000])
#ratios = [1, 2, 3, 4, 5]
```

```python
nheights = len(heights)
data = pandas.DataFrame({}, columns=['field', 'size', 'ratio', 'height', 'tesseroid_value'])
for size in sizes:
    dump_model(make_model(size), 'model.txt')
    print 'Calculating with tesseroids of size %g:' % (size)
    ntess = !wc -l model.txt
    print '  number of tesseroids: %s' % (ntess[0].split()[0])
    for ratio in ratios:
        # Assemble the pipeline
        if ratio == 0:
            flag = '-a'
        else:
            flag = '-t%g' % (ratio)
        commands = ["cat points.txt"]
        commands.extend('tess%s model.txt %s' % (field, flag) for field in fields)
        pipeline = " | ".join(commands)
        tmp = !$pipeline
        # Remove comments and convert the strings to floats
        output = np.loadtxt(StringIO('\n'.join(tmp)), unpack=True)[3:]
        for value, field in zip(output, fields):
            data = data.append(
                pandas.DataFrame(
                    {'field': field,
                     'size': size,
                     'ratio': ratio,
                     'height': heights,
                     'tesseroid_value': value,
                    }))
data[data.field == 'gz']
```
