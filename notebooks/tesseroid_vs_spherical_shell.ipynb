# A comparison of the effects of tesseroids against a spherical shell

This notebook compares the gravitational fields calculated for a tesseroid model against the analytical solution for a spherical shell.

The tesseroid fields are calculated using a range of distance/size ratios for the recursive division of the tesseroids. The sphell fields are used as a reference, or "ground truth". We use the difference between the tesseroid and reference fields to quantify the error in the tesseroid calculations for each distance/size ratio.

## Gravitational effect of a spherical shell

The analytical solution exists for an observation point located on the pole. We can pretend that the point is on the equator to avoid the numerical singularities of tesseroids at the pole.

The potential and its first and second derivatives are given by (Grombein et al., 2013):

$$
V(r) = \dfrac{4}{3}\pi G \rho \dfrac{r_2^3 - r_1^3}{r},
$$

$$
g_z(r) = \dfrac{V(r)}{r},
$$

$$
g_{zz}(r) = 2\dfrac{V(r)}{r^2},
$$

$$
g_{xx}(r) = g_{yy}(r) = -\dfrac{g_{zz}}{2},
$$


and (because of the symmetry of the shell)

$$
g_x(r) = g_y(r) = g_{xy}(r) = g_{xz}(r) = g_{yz}(r) = 0,
$$


where $\rho$ is the dentity, $r$ is the radius coordinate of the observation point, $r_1$ and $r_2$ are radial coordinates of the bottom and top of the spherical shell, respectively.

The function below calculates the shell fields.

```python
def calc_shell_effect(height, top, bottom, density):
    r = height + MEAN_EARTH_RADIUS
    # top and bottom are heights
    r1 = bottom + MEAN_EARTH_RADIUS
    r2 = top + MEAN_EARTH_RADIUS
    potential = (4/3)*np.pi*G*density*(r2**3 - r1**3)/r
    data = np.zeros(10)
    data[0] = potential
    data[3] = SI2MGAL*(potential/r)
    data[4] = SI2EOTVOS*(-potential/r**2)
    data[7] = SI2EOTVOS*(-potential/r**2)
    data[9] = SI2EOTVOS*(2*potential/r**2)
    return data
```

## Tesseroid model of the spherical shell

I'll use the `TesseroidMesh` of Fatiando a Terra to make a model of a spherical shell. The model needs to be written to a file.

The function below takes the size (in degrees), top, bottom and density, makes the tesseroids and saves them to a file.

```python
def make_model_file(size, top, bottom, density, verbose=False, return_mesh=False):
    bounds = [0, 360, -90, 90, top, bottom]
    nlon = int(360/size)
    assert nlon - 360/size == 0
    nlat = int(180/size)
    assert nlat - 180/size == 0
    mesh = TesseroidMesh(bounds, (1, nlat, nlon))
    # the 3rd dim is - because of a quirk in TesseroidMesh
    assert mesh.dims == (size, size, -(top - bottom))
    assert mesh.size == nlat*nlon
    fname = os.path.join(filedir, 'model-size{:g}.tess'.format(size))
    with open(fname, 'w') as f:
        for t in mesh:
            f.write('{} {} {} {} {} {} {}\n'.format(
                t.w, t.e, t.s, t.n, t.top, t.bottom, density))
    tmp = !cat $fname | grep -v "#" | grep -e "^$" -v| wc -l
    tess_in_file = int(tmp[0])
    assert mesh.size == tess_in_file
    if verbose:
        print('Model file: {}'.format(fname))
        print('Model size: {}'.format(tess_in_file))
        print('Head:')
        !head $fname
        print('Tail:')
        !tail $fname
    if return_mesh:
        return fname, mesh
    else:
        return fname
```

We'll make a 30x30 degree tesseroid model to test things.

```python
size = 30
top, bottom = 1000, 0
density = 2670
model_file, mesh = make_model_file(size, top, bottom, density, verbose=True, return_mesh=True)
```

Now we need to calculate the gravitational effect of this tesseroid model using a range of distance/size ratios. The function below takes a model file name,  height of observations, tesseroid size, and a distance to size ratio. It makes a computation grid at the given height. The grid is placed over a tesseroid and the area is the size of a tesseroid.

```python
def calc_tess_effect(fname, height, size, ratio):
    if ratio == 0:
        flag = '-a'
    else:
        flag = '-t{:f}'.format(ratio)
    tmp = 'effect-{}-ratio{:f}.txt'.format(os.path.split(fname)[1], ratio)
    outfile = os.path.join(filedir, tmp)
    # Make a computation grid above the tesseroid with corner at 0, 0
    lons, lats = gridder.regular([0, size]*2, (11, 11))
    grid_text = '\n'.join('{:f} {:f} {:f}'.format(lon, lat, height) 
                          for lon, lat in zip(lons, lats))
    !echo "$grid_text" | \
        tesspot $fname $flag | \
        tessgx $fname $flag | \
        tessgy $fname $flag | \
        tessgz $fname $flag | \
        tessgxx $fname $flag | \
        tessgxy $fname $flag | \
        tessgxz $fname $flag | \
        tessgyy $fname $flag | \
        tessgyz $fname $flag | \
        tessgzz $fname $flag > $outfile
    return outfile
```

Now we can specify a range of distance/size ratios and a computation height...

```python
ratio = np.arange(0, 7.5, 0.5)
height = top + 1000
```

... and calculate the effect of the tesseroid model for each ratio.

```python
tess_out_files = [calc_tess_effect(model_file, height, size, r) for r in ratio]
```

## Comparison of spherical shell effect and tesseroid model

Now I can load the calculated fields of the tesseroid model and compare it with the calculated effect for the real spherical shell.

```python
shell_data = calc_shell_effect(height, top, bottom, density)
shell_data
```

```python
def load_data(fname):
    return np.loadtxt(fname, usecols=range(3, 13))
```

The difference I'm interested in is the largest difference for each field (potential, gz, etc) for each distance/size ratio:

```python
def calc_differences(shell, tess_files):
    diff = np.empty((10, ratio.size))
    for i, fname in enumerate(tess_files):
        tess = load_data(fname)
        d = np.abs(shell - tess).max(axis=0)
        diff[:, i] = d
    return diff
```

```python
diff = calc_differences(shell_data, tess_out_files)
```

Now, plot the absolute and relative (in percentange) differences as a function of the distance/size ratio. The relative difference is the largest difference divided by the shell value.

```python
titles = [r'$V$', r'$g_{z}$', r'$g_{xx}$', r'$g_{yy}$', r'$g_{zz}$']
fields = [0, 3, 4, 7, 9]
fig, subplots = plt.subplots(5, 2, figsize=(10, 13))
for f, title, axes in zip(fields, titles, subplots):
    ax1, ax2 = axes
    ax1.text(0.8, 0.8, title, fontsize=18, transform=ax1.transAxes)
    ax1.plot(ratio, diff[f], '.-')
    ax1.set_yscale('log')
    ax1.grid(True)
    ax1.set_ylabel('Difference')
    ax2.plot(ratio, 100*diff[f]/np.abs(shell_data[f]), '.-')
    ax2.set_yscale('log')
    ax2.grid(True)
    ax2.set_ylabel('Difference (%)')
    ax2.hlines(0.5, ratio.min(), ratio.max(), colors='r')
ax1.set_xlabel('Distance/size ratio')
ax2.set_xlabel('Distance/size ratio')
ax1, ax2 = subplots[0]
ax1.set_title('Absolute difference')
ax2.set_title('Realtive difference')
plt.tight_layout()
```

## Repeat these computations various sizes

This is to check if the results are dependent on the size of the tesseroids.

```python
sizes = [5, 2, 1]
```

Make the various model files and calculate their effects:

```python
height = top + 1000
tess_files_per_size = []
for s in sizes:
    print('Calculating size {}'.format(s))
    mfile = make_model_file(s, top, bottom, density)
    tess_files_per_size.append([calc_tess_effect(mfile, height, s, r) for r in ratio])
```

Now plot the absolute and relative differences again. The red line marks a difference of 0.5%.

```python
titles = [r'$V$', r'$g_{z}$', r'$g_{xx}$', r'$g_{yy}$', r'$g_{zz}$']
fields = [0, 3, 4, 7, 9]
fig, subplots = plt.subplots(5, 2, figsize=(10, 13))
for s, tess_files in zip(sizes, tess_files_per_size): 
    shell = calc_shell_effect(height, top, bottom, density)
    diff = calc_differences(shell, tess_files)
    for f, title, axes in zip(fields, titles, subplots):
        ax1, ax2 = axes
        ax1.text(0.8, 0.8, title, fontsize=18, transform=ax1.transAxes)
        ax1.plot(ratio, diff[f], '.-', label=r'${:g}^\circ$'.format(s))
        ax1.set_yscale('log')
        ax1.grid(True)
        ax1.set_ylabel('Difference')
        ax2.plot(ratio, 100*diff[f]/np.abs(shell[f]), '.-', label=r'${:g}^\circ$'.format(s))
        ax2.set_yscale('log')
        ax2.grid(True)
        ax2.set_ylabel('Difference (%)')
for ax1, ax2 in subplots:
    ax2.hlines(0.5, ratio.min(), ratio.max(), colors='r')
    ax2.legend(loc='upper right', numpoints=1)
ax1.set_xlabel('Distance/size ratio')
ax2.set_xlabel('Distance/size ratio')
plt.tight_layout()
```

Since $g_x$ and $g_y$ and tensor the cross components are zero, I can't calculate the relative difference between the tesseroid model values and the shell values (which are zero). Instead, I'll calculate the difference relative to $g_z$ (for $g_x$ and $g_y$) and $g_{zz}$ (for $g_{xy}$, $g_{xz}$, and $g_{yz}$).

```python
titles = [r'$g_{x}$', r'$g_{y}$', r'$g_{xy}$', r'$g_{xz}$', r'$g_{yz}$']
fields = [1, 2, 5, 6, 8]
fig, subplots = plt.subplots(3, 2, sharex='col', sharey='row', figsize=(10, 7))
for s, tess_files in zip(sizes, tess_files_per_size): 
    shell = calc_shell_effect(height, top, bottom, density)
    diff = calc_differences(shell, tess_files)
    for f, title, ax in zip(fields, titles, subplots.ravel()):
        ax.text(0.5, 0.8, title, fontsize=18, transform=ax.transAxes)
        if f in [1, 2]:
            d = 100*diff[f]/np.abs(shell[3])
        else:
            d = 100*diff[f]/np.abs(shell[9])
        ax.plot(ratio, d, '.-', label=r'${:g}^\circ$'.format(s))
        ax.hlines(0.1, ratio.min(), ratio.max(), colors='r')
        ax.set_xlim(ratio.min(), ratio.max())
        ax.set_yscale('log')
        ax.grid(True)
        ax.set_yticks(ax.get_yticks()[2:-2])
for ax in subplots[:, 0]:
    ax.set_ylabel('Difference (%)')
ax.set_ylim(subplots[-2, 1].get_ybound())
ax.set_yticks(subplots[-2, 1].get_yticks())
subplots[-1, 0].set_xlabel('Distance/size ratio')
subplots[-1, 0].legend(loc='upper right', numpoints=1, bbox_to_anchor=(1.8, 0.8), fancybox=True, shadow=True)
lastax = subplots[-1, -1]
lastax.axison = False
plt.tight_layout()
plt.subplots_adjust(hspace=0, wspace=0)
```
